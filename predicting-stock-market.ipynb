```python
import sklearn 
from sklearn.model_selection import train_test_split
x=data[['OPEN','HIGH','LOW','LAST','PREVCLOSE','TOTTRDQTY','TOTALTRADES']]
y=data[['CLOSE']]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
print(np.shape(x_train), np.shape(y_train))
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion matrix, accuracy_score
```

```python
instance=  LinearRegression()
instance.fit(x_train,y_train)
print(instance.coef_)
print(instance.intercept_)
```

```python
pred=instance.predict(x_test)
print(x_test)
```
